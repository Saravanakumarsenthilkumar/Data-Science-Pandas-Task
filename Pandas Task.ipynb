```python
import pandas as pd
```

```python
df=pd.read_csv("Book1 (1).csv")
print(df)
```

```python
print(df.isnull().sum())
```

```python
print(df.dtypes)
```

```python
df['Age'] =df['Age'].fillna(df['Age'].mean())
print(df)
```

```python
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
print(df)
```

```python
filtered=df[df['Salary']>40000]
print(filtered)
```

```python
filtered=df[df['Department']=='Finance']
print(filtered)
```

```python
sorted_data = df.sort_values(by="Salary", ascending=False)
print(sorted_data)
```
