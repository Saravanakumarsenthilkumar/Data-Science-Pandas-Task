```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataframe.csv")
print(df)
```

```python
plt.plot(df["Study_Hours"], df["Marks"])
plt.xlabel("Study_Hours")
plt.ylabel("Marks")
plt.title("Student Score Analysis")
plt.legend(["Marks"])
plt.show()
```

```python
plt.plot(df["Study_Hours"],df["Marks"],color="violet",marker="*",markersize=16)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Student Score Analysis - Line Plot")
plt.legend(["Marks Trend"])
plt.show()
```

```python
plt.bar(df["Study_Hours"],df["Marks"],color="red")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Student Score Analysis - Bar Plot")
plt.legend(["Marks Comparison"])
plt.show()
```

```python
plt.hist(df["Marks"],bins=5,color="purple")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Marks Distribution")
plt.legend(["Marks Frequency"])
plt.show()
```

```python
plt.subplot(2,2,3)
plt.bar(df["Study_Hours"],df["Marks"],color="green")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Bar Plot Analysis")
plt.legend(["Marks Comparison"])

plt.show()
```

```python
plt.plot(df["Study_Hours"],df["Marks"],color="yellow",marker="x",linestyle="--")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Student Score Analysis")

plt.grid(True)

plt.legend(["Marks"])
plt.show()
```

```python
plt.plot(df["Study_Hours"],df["Marks"],color="green",marker="o")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Student Score Analysis")

plt.annotate("Highest Score",
             xy=(10,90),
             xytext=(7,75),
             arrowprops=dict(facecolor="orange"))

plt.legend(["Marks"])
plt.show()
```
