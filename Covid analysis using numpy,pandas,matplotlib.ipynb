```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(10)

days = np.arange(1,31)

confirmed = np.random.randint(1000,5000,30)
recovered = confirmed - np.random.randint(200,1000,30)
deaths = np.random.randint(10,100,30)

data = {
    "Day": days,
    "Confirmed": confirmed,
    "Recovered": recovered,
    "Deaths": deaths
}

df = pd.DataFrame(data)

print(df.head())
```

```python
print("Total Confirmed Cases:", df["Confirmed"].sum())
print("Total Recovered:", df["Recovered"].sum())
print("Total Deaths:", df["Deaths"].sum())

print("\nAverage Daily Cases")
print(df.mean())
```

```python
plt.plot(df["Day"], df["Confirmed"], marker="o", label="Confirmed")
plt.plot(df["Day"], df["Recovered"], marker="s", label="Recovered")
plt.plot(df["Day"], df["Deaths"], marker="*", label="Deaths")

plt.xlabel("Days")
plt.ylabel("Number of Cases")
plt.title("COVID-19 Trend Analysis")

plt.legend()
plt.grid()

plt.show()
```

```python
plt.bar(df["Day"], df["Confirmed"])

plt.xlabel("Day")
plt.ylabel("Confirmed Cases")
plt.title("Daily Confirmed Cases")

plt.show()
```

```python
totals = [
df["Confirmed"].sum(),
df["Recovered"].sum(),
df["Deaths"].sum()
]

labels = ["Confirmed","Recovered","Deaths"]

plt.pie(totals, labels=labels, autopct="%1.1f%%")

plt.title("COVID Case Distribution")

plt.show()
```

```python
plt.hist(df["Confirmed"], bins=8)

plt.xlabel("Confirmed Cases")
plt.ylabel("Frequency")
plt.title("Distribution of Confirmed Cases")

plt.show()
```
