```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataframe.csv")
print(df)
```

```python
plt.plot(df["Study_Hours"], df["Marks"], marker='o', color='blue')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.grid(True)
plt.show()
```

```python
plt.scatter(df["Study_Hours"], df["Marks"], color='red')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Scatter Plot of Study Hours vs Marks")
plt.show()
```

```python
plt.bar(df["Study_Hours"], df["Marks"], color='green')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Bar Chart of Study Hours vs Marks")
plt.show()
```

```python
plt.hist(df["Marks"], bins=5)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Marks Distribution")
plt.show()
```

```python
plt.pie(df["Marks"], labels=df["Study_Hours"], autopct='%1.1f%%')
plt.title("Marks Contribution by Study Hours")
plt.show()
```

```python
plt.boxplot(df["Marks"])
plt.title("Box Plot of Marks")
plt.ylabel("Marks")
plt.show()
```
