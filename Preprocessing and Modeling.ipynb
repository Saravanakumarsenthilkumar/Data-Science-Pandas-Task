```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Step 2: Create Dataset
data = {
    'Area': [1000, 1500, 1800, 2400, np.nan, 3000, 3500],
    'Bedrooms': [2, 3, 3, 4, 3, np.nan, 5],
    'Age': [10, 5, 8, 2, 7, 3, np.nan],
    'Price': [200000, 300000, 350000, 500000, 320000, 600000, 650000]
}

df = pd.DataFrame(data)

print("Original Dataset:\n", df)

# Step 3: Split Features & Target
X = df.drop('Price', axis=1)
y = df['Price']

# Step 4: Handle Missing Values (Imputation)
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Step 5: Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Step 6: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Step 7: Apply Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)

# Step 8: Prediction
y_pred = model.predict(X_test)

# Step 9: Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nPredictions:", y_pred)
print("Mean Squared Error:", mse)
print("R2 Score:", r2)

# Step 10: Model Coefficients
print("\nModel Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error, r2_score

# Step 2: Create Dataset
data = {
    'Area': [1000, 1500, 1800, 2400, np.nan, 3000, 3500],
    'Bedrooms': [2, 3, 3, 4, 3, np.nan, 5],
    'Age': [10, 5, 8, 2, 7, 3, np.nan],
    'Price': [200000, 300000, 350000, 500000, 320000, 600000, 650000]
}

df = pd.DataFrame(data)

# Step 3: Split Features & Target
X = df.drop('Price', axis=1)
y = df['Price']

# Step 4: Imputation
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Step 5: Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Step 6: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# -------------------------------
# 🔵 Model 1: Linear Regression
# -------------------------------
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

# -------------------------------
# 🌳 Model 2: Decision Tree
# -------------------------------
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

# -------------------------------
# 🌲 Model 3: Random Forest
# -------------------------------
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# Step 7: Evaluation Function
def evaluate(model_name, y_test, y_pred):
    print(f"\n{model_name}")
    print("MSE:", mean_squared_error(y_test, y_pred))
    print("R2 Score:", r2_score(y_test, y_pred))

# Step 8: Evaluate All Models
evaluate("Linear Regression", y_test, y_pred_lr)
evaluate("Decision Tree", y_test, y_pred_dt)
evaluate("Random Forest", y_test, y_pred_rf)
```
