```python
import pandas as pd
dataframe=pd.read_csv("data.csv")
print(dataframe)
```

```python
#GroupBy Department wise Total Salary
avg=dataframe.groupby("Department")["Salary"].sum()
print(avg)
```

```python
#Department wise total Employee using Groupby
count_employee=dataframe.groupby("Department")["Name"].count()
print(count_employee)
```

```python
#Read New Data for Merging Operation:
incentive_dataframe=pd.read_csv("incentive_dataframe.csv")
print(incentive_dataframe)
```

```python
#Merging Two table based "Name"
merged_data=pd.merge(dataframe,incentive_dataframe,on="Name")
print(merged_data)
```

```python
#Concatination of Two Data
concat=pd.concat([dataframe,incentive_dataframe])
print(concat)
```

```python
#Correlation Analysis
correlation_analysis=dataframe[["JoinDate","Salary"]].corr()
print(correlation_analysis)
```
