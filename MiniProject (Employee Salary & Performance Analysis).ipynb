```python
import numpy as np
import pandas as pd

```

```python
np.random.seed(42)
names = ["Arun","Priya","Karthik","Meena","Rahul","Divya","Ajay","Sneha","Vijay","Anu"]

departments = ["IT","HR","Finance","Sales","Marketing"]

# Generate dataset using numpy
emp_id = np.arange(101,131)
name_col = np.random.choice(names,30)
dept_col = np.random.choice(departments,30)
experience = np.random.randint(1,10,30)
salary = np.random.randint(30000,80000,30)
performance = np.random.randint(60,100,30)

# Incentive calculation
incentive = performance * 100

# Create DataFrame
df = pd.DataFrame({
    "EmpID":emp_id,
    "Name":name_col,
    "Department":dept_col,
    "Experience":experience,
    "Salary":salary,
    "PerformanceScore":performance,
    "Incentive":incentive
})

# Save to Excel
df.to_excel("employee_data.xlsx",index=False)

print("Dataset created successfully!")
print(df.head())
```

```python
# Average salary by department
print("\nAverage Salary by Department")
print(df.groupby("Department")["Salary"].mean())

# Top 5 highest paid employees
print("\nTop 5 Salaries")
print(df.nlargest(5,"Salary"))

# Correlation between experience and salary
print("\nCorrelation")
print(df[["Experience","Salary"]].corr())

# Best performing employee
print("\nBest Performer")
print(df.loc[df["PerformanceScore"].idxmax()])
```

```python
df["Salary_Level"] = pd.cut(df["Salary"],
                           bins=[30000,45000,60000,80000],
                           labels=["Low","Medium","High"])

print(df[["Name","Salary","Salary_Level"]])
```
